```python
import os
import argparse
import time

import torch
import torch.nn.parallel
import torch.optim
import torch.utils.data.distributed

from src_files.helper_functions.bn_fusion import fuse_bn_recursively
from src_files.models import create_model
import matplotlib

from src_files.models.tresnet.tresnet import InplacABN_to_ABN

#matplotlib.use('TkAgg')
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

from types import SimpleNamespace

import pandas as pd

pred_cars = pred_boxes[(pred_boxes["class"] == 2) | (pred_boxes["class"] == 7)]


image_path = '../datasets/datasets_test/test/'

d_args = {'num_classes': 196, 'model_path': 'tresnet_l_stanford_card_96.41.pth', 'pic_path': '../datasets/datasets_train/train/15007024230.jpg', 'model_name': 'tresnet_l', 'image_size': 384, 'th': 0.95, 'top_k': 20, 'use_ml_decoder': 1, 'num_of_groups': -1, 'decoder_embedding': 768, 'zsl': 0}
args = SimpleNamespace(**d_args)

results = []
model_names = []

print('Inference code on a single image')


# Setup model
print('creating model {}...'.format(args.model_name))
model = create_model(args, load_head=True).cuda()
state = torch.load(args.model_path, map_location='cpu')
model.load_state_dict(state['model'], strict=True)
########### eliminate BN for faster inference ###########
model = model.cpu()
model = InplacABN_to_ABN(model)
model = fuse_bn_recursively(model)
model = model.cuda().half().eval()
#######################################################
print('done')


classes_list = np.array(list(state['idx_to_class']))
print('done\n')

# doing inference
print('loading image and doing inference...')


for index, row in pred_cars.iterrows():
    try:
        im = Image.open(image_path+row['im_name'])
        plt.imshow(im)
        im_resize = im.resize((args.image_size, args.image_size))
        np_img = np.array(im_resize, dtype=np.uint8)
        try:
            tensor_img = torch.from_numpy(np_img).permute(2, 0, 1).float() / 255.0  # HWC to CHW
        except:
            print(row['im_name'],'failed')
            continue
        tensor_batch = torch.unsqueeze(tensor_img, 0).cuda().half() # float16 inference
        output = torch.squeeze(torch.sigmoid(model(tensor_batch)))
        np_output = output.cpu().detach().numpy()


        ## Top-k predictions
        # detected_classes = classes_list[np_output > args.th]
        idx_sort = np.argsort(-np_output)
        detected_classes = np.array(classes_list)[idx_sort][: args.top_k]
        scores = np_output[idx_sort][: args.top_k]
        idx_th = scores > args.th
        detected_classes = detected_classes[idx_th]

        model_name = {'im_name': row['im_name'], 'pred_models': detected_classes}
        model_names.append(model_name)
        print(row['im_name'],detected_classes)
    except Error as e:
        model_name = {'im_name': row['im_name'], 'pred_models': None}
        model_names.append(model_name)
        print("error",row['im_name'])
        

print('final done\n')





```

```python
carbon = pd.read_csv('datasets/car_models_footprint.csv',sep=';')
```

```python
carbon.head()
```

```python
from difflib import SequenceMatcher
import math 
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

carbon_emission = []

ref_models = carbon["models"].values.tolist()

for index, row in models.iterrows():
    pred_models = row["pred_models"]
    best_pred = ''
    best_score = 0
    if type(pred_models) == float:
        continue
    if pred_models is None:
        carbon_emission.append({'im_name': row['im_name'], 'e': 0})
        continue
    for model in pred_models:

            
        for ref in ref_models:
            score = similar(ref,model.replace('_', ' '))             
            if score > best_score:
                      best_score = score
                      best_pred = ref
    carbon_km = carbon[carbon["models"]==str(best_pred)]["Average of CO2 (g per km)"].values.tolist()
    
    if len(carbon_km) > 0:
        carbon_emission.append({'im_name': row['im_name'], 'e': float(carbon_km[0])})
    else:
        carbon_emission.append({'im_name': row['im_name'], 'e': 0})
```
