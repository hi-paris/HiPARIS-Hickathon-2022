# Resnet

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

#Librairie de path
import pathlib
import pandas as pd
```

```python
number_class=30
dataset = "car_models_database_train/"
```

```python
BATCH_SIZE = 32
IMG_H = 224
IMG_W = 224
```

```python
    TFs = {"horizontal_flip": True,
        "vertical_flip": True,
        "rotation_range": 30,
        "brightness_range": (1, 2)
        }

    datagen = ImageDataGenerator(**TFs)
```

```python
    train_generator = datagen.flow_from_directory(
        dataset,
        color_mode="rgb",
        batch_size=BATCH_SIZE,
        subset = "training",
        target_size=(IMG_H,IMG_W),
        shuffle=True,
        class_mode='categorical')

    validation_generator = datagen.flow_from_directory(
        dataset,
        color_mode="rgb",
        batch_size=BATCH_SIZE,
        subset = "validation",
        target_size=(IMG_H,IMG_W),
        shuffle=True,
        class_mode='categorical')
```

```python
train_generator.class_indices
```

```python
train_generator.image_shape
tf.config.list_physical_devices('GPU')
tf.debugging.set_log_device_placement(True)
```

```python
model_resnet = tf.keras.applications.resnet50.ResNet50(
    include_top=True, weights=None, input_tensor=None,
    input_shape=None, pooling=None, classes=31, classifier_activation='softmax'
)
```

```python
model_resnet.compile(optimizer="Adam", 
        loss=tf.keras.losses.categorical_crossentropy,
        metrics=['accuracy'])
```

```python
model_resnet.compile(optimizer="Adam", 
        loss=tf.keras.losses.sparse_categorical_crossentropy,
        metrics=['accuracy', 'sparse_top_k_categorical_accuracy'])
```

```python
history = model_resnet.fit(train_generator, validation_data=validation_generator,epochs=15,batch_size=BATCH_SIZE, verbose=1,validation_steps=1, workers=4)
```

```python
model_resnet.save('resnet50_30epochs.h5')
```

```python
import matplotlib.pyplot as plt
plt.subplot(131)
plt.plot(history.history['accuracy']    ,'--',     label='accuracy', c='g')
plt.plot(history.history['val_accuracy'],     label = 'val_accuracy', c='g')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.0, 1.05])
plt.subplot(132)
plt.plot(history.history['sparse_top_k_categorical_accuracy']  ,'--'  ,      label='accuracy')
plt.plot(history.history['val_sparse_top_k_categorical_accuracy'],      label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy top-5')
plt.ylim([0.0, 1.05])
plt.subplot(133)
plt.plot(history.history['loss']    ,'--',     label='loss')
plt.plot(history.history['val_loss'],     label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
```
